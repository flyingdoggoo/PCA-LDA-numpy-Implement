```python
means = []
for col in range(4):
    col_sum = 0
    for row in X:
        col_sum += row[col]
    mean = col_sum / len(X)
    means.append(mean)
print("Means: ", means)
```

```python
stds = []
for col in range(4):
    sum = 0
    for row in X:
        sum += (row[col] - means[col])**2
    sum /= len(X)
    std = sum ** 0.5
    stds.append(std)
print("Standard Deviation: ", stds)
```

```python
def standardization(X, means, stds):
    for col in range(4):
        for row in X:
            row[col] = (row[col] - means[col])/stds[col]
```

```python
def standardize(X):
    mean = np.mean(X, axis = 0)
    X_centered = X - mean
    return X_centered
def covariance_matrix(X):
    return np.cov(X.T)
    #MT hiệp phương sai
def eigen_decomposition(C):
    eigenvalues, eigenvectors = np.linalg.eig(C)
    #Giá trị riêng, vecto riêng
    return eigenvalues, eigenvectors
def select_principal_components(eigenvalues, eigenvectors, n_components=2):
    idx = np.argsort(eigenvalues)[::-1]
    selected_vectors = eigenvectors[:, idx[:n_components]]
    return selected_vectors
def pca_transform(X, components):
    return X.dot(components)
def pca_manual(X, n_components=2):
    X_centered = standardize(X)
    cov_matrix = covariance_matrix(X_centered)
    eigenvalues, eigenvectors = eigen_decomposition(cov_matrix)
    components = select_principal_components(eigenvalues, eigenvectors, n_components)
    X_pca = pca_transform(X_centered, components)
    return X_pca
```

```python
X_pca_manual = pca_manual(X, 2)
X_pca_manual[:, 1] *= -1
plot_pca(X_pca_manual, y_numeric, 'PCA thủ công')
```

```python
import numpy as np

# Hàm tính vector kỳ vọng (mean vector) của mỗi class
def mean_vectors(X, y):
    class_labels = np.unique(y)
    mean_vecs = []
    for label in class_labels:
        mean_vecs.append(np.mean(X[y == label], axis=0))
    return mean_vecs

# Hàm tính ma trận scatter trong-class (Within-class scatter matrix)
def within_class_scatter(X, y, mean_vecs):
    n_features = X.shape[1]
    S_w = np.zeros((n_features, n_features))  # Ma trận scatter trong-class
    for label, mean_vec in zip(np.unique(y), mean_vecs):
        class_scatter = np.zeros((n_features, n_features))  # Ma trận scatter của từng class
        for row in X[y == label]:
            row, mean_vec = row.reshape(n_features, 1), mean_vec.reshape(n_features, 1)
            class_scatter += (row - mean_vec).dot((row - mean_vec).T)
        S_w += class_scatter  # Tổng hợp scatter trong-class của các class
    return S_w

# Hàm tính ma trận scatter giữa-class (Between-class scatter matrix)
def between_class_scatter(X, y, mean_vecs):
    overall_mean = np.mean(X, axis=0)
    n_features = X.shape[1]
    S_b = np.zeros((n_features, n_features))  # Ma trận scatter giữa-class
    for i, mean_vec in enumerate(mean_vecs):
        n = X[y == i].shape[0]
        mean_vec = mean_vec.reshape(n_features, 1)
        overall_mean = overall_mean.reshape(n_features, 1)
        S_b += n * (mean_vec - overall_mean).dot((mean_vec - overall_mean).T)
    return S_b

# LDA projection
def lda(X, y):
    mean_vecs = mean_vectors(X, y)
    S_w = within_class_scatter(X, y, mean_vecs)
    S_b = between_class_scatter(X, y, mean_vecs)
    
    # Tính các eigenvalue và eigenvector từ ma trận S_w^-1 S_b
    eig_vals, eig_vecs = np.linalg.eig(np.linalg.inv(S_w).dot(S_b))
    
    # Sắp xếp eigenvalue theo thứ tự giảm dần và chọn các eigenvector tương ứng
    eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:, i]) for i in range(len(eig_vals))]
    eig_pairs = sorted(eig_pairs, key=lambda x: x[0], reverse=True)
    
    # Lấy eigenvector ứng với eigenvalue lớn nhất
    W = np.hstack([eig_pairs[i][1].reshape(X.shape[1], 1) for i in range(1)])  # Vector chiếu (projection vector)
    
    # Chiếu dữ liệu lên không gian mới
    X_lda = X.dot(W)
    
    return X_lda, W

# Áp dụng LDA
X_lda, W = lda(X, y_numeric)

print("Dữ liệu sau khi chiếu LDA:\n", X_lda)
print("Vector chiếu W:\n", W)
```

```python
def plot_lda(X_lda, y):
    plt.figure(figsize=(8, 6))
    for label in np.unique(y):
        plt.scatter(X_lda[y == label], np.zeros_like(X_lda[y == label]), label=f'Class {label}')
    
    plt.xlabel('LDA Component 1')
    plt.title('LDA Projection')
    plt.legend()
    plt.grid(True)
    plt.show()

# Gọi hàm vẽ
plot_lda(X_lda, y)
```
