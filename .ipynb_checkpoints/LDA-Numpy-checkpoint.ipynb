```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = load_iris()
```

```python
X, y = data.data, data.target
```

```python
def normalize(X):
    X = (X - np.mean(X, axis = 0)) / (np.std(X, axis = 0))
    return X
```

```python
X = normalize(X)
```

```python
dim_name = ['Chiều dài đài hoa', 'Chiều rộng đài hoa', 'Chiều dài cánh hoa', 'Chiều rộng cánh hoa']
```

```python
def plot(X, y, dim1, dim2, title):
    plt.scatter(X[:, dim1], X[:, dim2], c = y, cmap='viridis')
    if title == 'LDA':
        plt.xlabel('Thành phần chính 1')
        plt.ylabel('Thành phần chính 2')
    else:
        plt.xlabel(dim_name[dim1])
        plt.ylabel(dim_name[dim2])
    plt.grid()
    plt.show()
```

```python
plot(X, y, 0, 1, 'No LDA')
```

```python
def LDA(X, y, n_compos = 2):
    #We need 150x4 x 4x2 -> thats the eigenvectors vectors shape: 4x2
    idx = np.unique(y)
    mean_all = np.mean(X, axis = 0)
    S_w = np.zeros((X.shape[1], X.shape[1]))
    S_b = np.zeros((X.shape[1], X.shape[1]))
    for c in idx:
        X_c = X[y == c]
        mean = np.mean(X_c)
        S_w += (X_c - mean).T.dot(X_c - mean)
        n = X_c.shape[0]
        S_b += (X_c - mean_all).T.dot(X_c - mean_all) * n
        # Sw^-1Sb
    mat = np.linalg.inv(S_w).dot(S_b)
    eigenvalues, eigenvectors = np.linalg.eig(mat)
    eigenvectors = eigenvectors.T
    idx = np.argsort(eigenvalues)[::-1]
    eigenvectors = eigenvectors[0:n_compos]
    eigenvalues = eigenvalues[0:n_compos]
    return np.dot(X, eigenvectors.T)
```
