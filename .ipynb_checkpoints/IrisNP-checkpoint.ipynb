```python
X = [row[:4] for row in data]
y = [row[4] for row in data]
saved = [row[:4] for row in data]
```

```python
means = []
for col in range(4):
    col_sum = 0
    for row in X:
        col_sum += row[col]
    mean = col_sum / len(X)
    means.append(mean)
print("Means: ", means)
```

```python
stds = []
for col in range(4):
    sum = 0
    for row in X:
        sum += (row[col] - means[col])**2
    sum /= len(X)
    std = sum ** 0.5
    stds.append(std)
print("Standard Deviation: ", stds)
```

```python
def standardization(X, means, stds):
    for col in range(4):
        for row in X:
            row[col] = (row[col] - means[col])/stds[col]
```

```python
standardization(X, means, stds)
```

```python
print("Before Standardization: ", saved[1])
print("After Standardization: ", X[1])
```

```python
y = np.array(y)
X = np.array(X)
saved = np.array(saved)
```

```python
flower_names = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
y_numeric = []
for name in y:
    index = flower_names.index(name)
    y_numeric.append(index)
```

```python
colors = ['r', 'g', 'b']
dim = ['Chiều dài đài hoa', 'Chiều rộng đài hoa', 'Chiều dài cánh hoa', 'Chiều rộng cánh hoa']
```

```python
def plot_data(X, y, dim1, dim2, title):
    plt.scatter(X[:, dim1], X[:, dim2], c = y)
    plt.xlabel(dim[dim1])
    plt.ylabel(dim[dim2])
    plt.title(title)
    plt.grid()
    plt.show()
```

```python
plot_data(X, y_numeric, 1, 2, 'Chưa chuẩn hóa')
plot_data(saved, y_numeric, 1, 2, 'Đã chuẩn hóa')
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
def plot_pca(X_pca, y, title):
    plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
    plt.xlabel('Chiều 1')
    plt.ylabel('Chiều 2')
    plt.title(title)
    plt.colorbar()
    plt.grid()
    plt.show()
plot_pca(X_pca, y_numeric, 'PCA thư viện')
```

## Các Bước

```python
def standardize(X):
    mean = np.mean(X, axis = 0)
    X_centered = X - mean
    return X_centered
def covariance_matrix(X):
    return np.cov(X.T)
    #MT hiệp phương sai
def eigen_decomposition(C):
    eigenvalues, eigenvectors = np.linalg.eig(C)
    #Giá trị riêng, vecto riêng
    return eigenvalues, eigenvectors
def select_principal_components(eigenvalues, eigenvectors, n_components=2):
    idx = np.argsort(eigenvalues)[::-1]
    selected_vectors = eigenvectors[:, idx[:n_components]]
    return selected_vectors
def pca_transform(X, components):
    return X.dot(components)
def pca_manual(X, n_components=2):
    X_centered = standardize(X)
    cov_matrix = covariance_matrix(X_centered)
    eigenvalues, eigenvectors = eigen_decomposition(cov_matrix)
    components = select_principal_components(eigenvalues, eigenvectors, n_components)
    X_pca = pca_transform(X_centered, components)
    return X_pca
```

```python
X_pca_manual = pca_manual(X, 2)
X_pca_manual[:, 1] *= -1
plot_pca(X_pca_manual, y_numeric, 'PCA thủ công')
```
